```python
import pandas as pd
```

```python
# Load & drop unnamed index col (first col)
df_train = pd.read_csv('../data/fraudtrain.csv', index_col=0)  # Magic: Makes unnamed_0 the index (discards it)
df_test = pd.read_csv('../data/fraudtest.csv', index_col=0)
```

```python
# Clean types
df_train['trans_date_trans_time'] = pd.to_datetime(df_train['trans_date_trans_time'])
df_train['dob'] = pd.to_datetime(df_train['dob'])
df_train['is_fraud'] = df_train['is_fraud'].astype(bool)
df_train['amt'] = pd.to_numeric(df_train['amt'], errors='coerce')

# Same for test
df_test['trans_date_trans_time'] = pd.to_datetime(df_test['trans_date_trans_time'])
df_test['dob'] = pd.to_datetime(df_test['dob'])
df_test['is_fraud'] = df_test['is_fraud'].astype(bool)
df_test['amt'] = pd.to_numeric(df_test['amt'], errors='coerce')
```

```python
# Combine & save cleaned (subset optional)
df = pd.concat([df_train, df_test])
df = df.sample(frac=0.5, random_state=42)  # ~750k rows for Supabase; remove for full
df.to_csv('../data/transactions_clean.csv', index=False)
```
