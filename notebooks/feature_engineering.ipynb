```python
import pandas as pd

df = pd.read_csv('../data/transactions_clean.csv')
df.head()
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])

df['transaction_hour'] = df['trans_date_trans_time'].dt.hour
df['transaction_day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
df['transaction_month'] = df['trans_date_trans_time'].dt.month

df.head()
```

```python
#Calculate Age
#only considering years for simplicity
df['age'] = (df['trans_date_trans_time'].dt.year - df['dob'].dt.year)
df.head()
```

```python
# Off-hours flag (midnight to 5am)
df['is_off_hours'] = df['transaction_hour'].between(0, 4).astype(int)
```

```python
# Category risk weighting
category_risk = {
    'travel': 3,
    'cash_advance': 3,
    'shopping_net': 2,
    'shopping_pos': 2,
    'entertainment': 2,
    'food_dining': 1,
    'gas_transport': 1,
    'grocery_pos': 1,
    'grocery_net': 1,
    'health_fitness': 1,
    'home': 1,
    'kids_pets': 1,
    'misc_net': 2,
    'misc_pos': 1,
    'personal_care': 1
}
df['category_risk_weight'] = df['category'].map(category_risk).fillna(1)
```

```python
#Monthly Spend
monthly = df.groupby(['cc_num', 'transaction_month'])['amt'].sum().reset_index()
monthly.columns = ['cc_num', 'transaction_month', 'monthly_spend']
df = df.merge(monthly, on=['cc_num', 'transaction_month'], how='left')
df.head()
```

```python
df.to_csv('../data/transactions_features.csv', index=False)
print(f"Saved {df.shape[0]} rows, {df.shape[1]} columns")
```
